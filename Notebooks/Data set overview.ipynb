```python
df['Soil_type'] = df['Soil_type'].astype(str)
```

```python
df['crop_type'] = df['crop_type'].astype(str)
```

```python
dfDummies = pd.get_dummies(df['Soil_type'], prefix = 'Soil')
```

```python
df = pd.concat([df,dfDummies],axis = 1)
```

```python
dfDummies1 = pd.get_dummies(df['crop_type'], prefix = 'Crop')
df = pd.concat([df,dfDummies1],axis = 1)
```
